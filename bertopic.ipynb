```python
import pandas as pd 
import numpy as np
from bertopic import BERTopic
import re

#load data 
import pandas as pd 
 
constitution_data = pd.read_csv("./data_full_fixed.csv", engine='python',sep='\t')

docs= []

valueList = constitution_data.values.tolist()
for value in valueList:
    #value[3] = re.sub(r'<span.*?>', '', value[3])
    #value[3] = re.sub(r'<\/span>', '', value[3])
    #display(value)
    docs.append(value[3])

#<span.*?> TODO remove this regex
display(docs)
```

```python
from transformers.pipelines import pipeline
from sentence_transformers import SentenceTransformer
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords 
pt_stopwords = stopwords.words('portuguese')
#pt_stopwords.extend(['caput','art','inciso','artigo','i','ii','iii','iv','incisos','167'])


# rufimelo/Legal-BERTimbau-large-TSDAE-v4-GPL-sts <- funciona
# rufimelo/Legal-BERTimbau-sts-large-ma-v3 <- não funciona
# stjiris/bert-large-portuguese-cased-legal-tsdae-gpl-nli-sts-v0 <- extremamente lento
# stjiris/bert-large-portuguese-cased-legal-tsdae-gpl-nli-sts-MetaKD-v1 <- mais rápido, tem um tópico 000 0000 ?


# we add this to remove stopwords
vectorizer_model = CountVectorizer(strip_accents='ascii',ngram_range=(2, 2),stop_words=pt_stopwords)
#vectorizer_model = CountVectorizer(stop_words=pt_stopwords)
#embedding_model = pipeline("feature-extraction", model="ulysses-camara/legal-bert-pt-br")



sentence_model = SentenceTransformer("stjiris/bert-large-portuguese-cased-legal-tsdae-gpl-nli-sts-MetaKD-v1")
model = BERTopic(embedding_model=sentence_model,
                 vectorizer_model=vectorizer_model,
                 calculate_probabilities=True,
                 language='portuguese')

topics, probabilities = model.fit_transform(docs)
```

```python
from scipy.cluster import hierarchy as sch
# Hierarchical topics
linkage_function = lambda x: sch.linkage(x, 'single', optimal_ordering=True)
hierarchical_topics = model.hierarchical_topics(docs, linkage_function=linkage_function)
model.visualize_hierarchy(hierarchical_topics=hierarchical_topics,orientation='bottom')
```

```python
from sentence_transformers import SentenceTransformer

#sentence_model = SentenceTransformer("ulysses-camara/legal-bert-pt-br")
embeddings = sentence_model.encode(docs, show_progress_bar=False)


model.visualize_hierarchical_documents(docs, hierarchical_topics, embeddings=embeddings)
```
